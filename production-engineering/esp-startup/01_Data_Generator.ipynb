```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

n_starts = 200
data = []

for i in range(n_starts):
    start_id = f"START_{i:03d}"
    
    # Randomly decide if this is a "Normal" (0) or "Hard" (1) start
    is_hard = np.random.choice([0, 1], p=[0.8, 0.2])
    
    if is_hard == 0:
        # Normal Start parameters
        peak_amp = np.random.normal(60, 2)       # Normal peak
        steady_amp = np.random.normal(50, 1)     # Steady state
        ramp_time = np.random.normal(5, 1)       # Fast ramp up
        vibration = np.random.normal(0.2, 0.05)  # Low vibration
    else:
        # Hard Start parameters
        peak_amp = np.random.normal(85, 5)       # High spike
        steady_amp = np.random.normal(52, 2)     # Slightly erratic steady
        ramp_time = np.random.normal(15, 3)      # Slow/Struggling ramp
        vibration = np.random.normal(0.8, 0.2)   # High vibration
        
    data.append([start_id, peak_amp, steady_amp, ramp_time, vibration])

df = pd.DataFrame(data, columns=['start_id', 'peak_amps', 'steady_amps', 'ramp_up_time_sec', 'max_vibration_g'])
df.to_csv('esp_startup_data.csv', index=False)
print("CSV Generated!")
```

```python
import pandas as pd
import numpy as np

# Configuration
n_injectors = 10
n_producers = 40
avg_inj_rate = 5000  # bbl/d
avg_prod_rate = 1500 # bbl/d
injected_tracer_conc = 1000 # ppm (Constant concentration for injectors)

# Generate Injectors Data
injectors = pd.DataFrame({
    'Well_ID': [f'INJ-{i:02d}' for i in range(1, n_injectors + 1)],
    'Type': 'Injector',
    'Water_Rate_bpd': np.random.normal(avg_inj_rate, 500, n_injectors).round(2),
    'Tracer_Conc_ppm': injected_tracer_conc # Constant injection concentration
})

# Generate Producers Data
# We simulate varying degrees of breakthrough (0 ppm to 900 ppm)
producers = pd.DataFrame({
    'Well_ID': [f'PROD-{i:02d}' for i in range(1, n_producers + 1)],
    'Type': 'Producer',
    'Water_Rate_bpd': np.random.normal(avg_prod_rate, 300, n_producers).round(2),
    'Tracer_Conc_ppm': np.random.uniform(0, 950, n_producers).round(2)
})

# Combine and Save
df = pd.concat([injectors, producers])
df.to_csv('water_recirculation_data.csv', index=False)
print("CSV file 'water_recirculation_data.csv' generated successfully.")
```
